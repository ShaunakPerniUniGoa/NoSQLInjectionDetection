```python
import csv

def read_csv(filename):
    data = {}
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)
        for row in reader:
            metric = row[0]
            for i in range(1, len(row)):
                if headers[i] not in data:
                    data[headers[i]] = {}
                if metric not in data[headers[i]]:
                    try:
                        data[headers[i]][metric] = float(row[i])
                    except ValueError:
                        data[headers[i]][metric] = row[i]  # If not convertible to float, store as string
    return data

def find_best_models(data):
    best_models = {}
    for dataset, metrics in data.items():
        best_models[dataset] = {}
        for metric, scores in metrics.items():
            if isinstance(scores, dict):
                best_model = max(scores, key=scores.get)
                best_models[dataset][metric] = (best_model, scores[best_model])
            else:
                best_models[dataset][metric] = (None, scores)  # If only one score, no best model
    return best_models


filename = './metrics.csv'
data = read_csv(filename)
best_models = find_best_models(data)
for dataset, metrics in best_models.items():
    print(f"Dataset {dataset}:")
    for metric, (best_model, score) in metrics.items():
        print(f"- Best {metric}: {best_model} ({score})")
```
